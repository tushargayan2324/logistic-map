```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def f_map(r,x_0):
    x = x_0
    for i in range(1000):
        x = r*(x)*(1-x)
    return np.array(x)
```

```python
r = np.linspace(3.5,4,2000)

X = np.linspace(0,1,200)

Y = []

for p in r:
    Y.append(f_map(p,X))
```

```python
plt.rcParams['figure.figsize'] = [26, 14]

plt.style.use('dark_background')

for xe, ye in zip(r,Y):
    plt.scatter( [xe] * len(ye), ye, s=0.01,color='white')

#plt.ylim(0.3,0.7)
#plt.xticks([i for i in range(10)])
#plt.axes().set_xticklabels([str(i) for i in range()])
```
